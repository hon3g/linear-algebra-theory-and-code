### Vector addtion

```python
# use np.array for vector math operations
v1 = np.array([3, -1])
v2 = np.array([2,  4])

v3 = v1 + v2

plt.plot((0, v1[0]), (0, v1[1]), 'b', label='v1')
plt.plot((0, v2[0])+v1[0], (0, v2[1])+v1[1], 'r', label='v2')
plt.plot((0, v3[0]), (0, v3[1]), 'k', label='v1+v2')
plt.legend()
plt.axis('square')
plt.axis((-6, 6, -6, 6))
plt.grid()
plt.title('v1 + v2\n')
plt.show()
```

### Vector * scalar

### Vector dot product

```python
# many ways to compute the dot product

v1 = np.array([1,  2, 3,  4, 5, 6])
v2 = np.array([0, -4, 1, -3, 6, 5])

dp1 = sum(np.multiply(v1, v2))

dp2 = np.dot(v1, v2)

dp3 = np.matmul(v1, v2)

dp4 = 0
for i in range(len(v1)):
    dp4 += v1[i] * v2[i]
    
print(dp1, dp2, dp3, dp4)
```

### Dot product properties

```python
# create 3 random vectors
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)
```

```python
# distributive
res1 = np.dot(a, (b+c))
res2 = np.dot(a, b) + np.dot(a, c)

print(res1, res2)
```

```python
# not associative
# one zeros vector
# a==b==c

# create 3 random vectors
n = 5
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)


bc = np.dot(b, c)
res1 = np.dot(a, bc)

ab = np.dot(a, b)
res2 = np.dot(ab, c)

print(res1, res2, sep='\n')
```

### The dot product: geometric perspective

```python
v1 = np.array([2,  4, -3])
v2 = np.array([0, -3, -3])

# compute the angle(radians) between two vectors
ang = np.arccos(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))
print(ang)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot((0, v1[0]), (0, v1[1]), (0, v1[2]), 'b')
ax.plot((0, v2[0]), (0, v2[1]), (0, v2[2]), 'r')
plt.axis((6, -6, 6, -6))
plt.title(f'The angle between the vectors is {round(ang, 2)} rad.')
plt.show()
```

```python
# algebraic and geometric dot product formulas

v1 = np.array([2,  4, -3])
v2 = np.array([0, -3, -3])

# algebraic
dp_a = np.dot(v1, v2)

# geometric
dp_b = np.linalg.norm(v1) * np.linalg.norm(v2) * np.cos(ang)

print(dp_a)
print(dp_b)
```

### Hermitain/Conjugate transpose

```python
# Conjugate fot complex num
z = np.complex(3, 4)
print(z)

# magnitude
print( np.linalg.norm(z) )

# regular transpose z ^T z
print( np.transpose(z)*z )

# conjugate transpose z ^H z
print( np.transpose(z.conjugate())*z )
```

```python
v = np.array( [3, 4j, 5+2j, np.complex(2,-5)] )

print(v.T)
print(np.transpose(v))
print('-----------------------------')
print( np.transpose(v.conjugate()) )
```

### Unit Vector

```python
# ||mu vector|| = 1

# vector
v1 = np.array([-3, 5])
# mu
mu = 1/np.linalg.norm(v1)

v1n = v1*mu

# unit vector in the same direction as v1
plt.plot((0, v1n[0]), (0, v1n[1]), 'r', label='v1-norm', linewidth=5)
plt.plot((0, v1[0]), (0, v1[1]), 'b', label='v1')
plt.grid()
plt.legend()
plt.axis('square')
plt.axis((-6, 6,  -6, 6))
```

```python
# ||mu vector|| = 1

# vector
v1 = np.array([-3, 5]) / 10
# mu
mu = 1/np.linalg.norm(v1)

v1n = v1*mu

plt.plot((0, v1n[0]), (0, v1n[1]), 'r', label='v1-norm', linewidth=5)
plt.plot((0, v1[0]), (0, v1[1]), 'b', label='v1')
plt.grid()
plt.legend()
plt.axis('square')
plt.axis((-6, 6,  -6, 6))
```
