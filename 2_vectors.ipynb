### Vector addtion

```python
# use np.array for vector math operations
v1 = np.array([3, -1])
v2 = np.array([2,  4])

v3 = v1 + v2

plt.plot((0, v1[0]), (0, v1[1]), 'b', label='v1')
plt.plot((0, v2[0])+v1[0], (0, v2[1])+v1[1], 'r', label='v2')
plt.plot((0, v3[0]), (0, v3[1]), 'k', label='v1+v2')
plt.legend()
plt.axis('square')
plt.axis((-6, 6, -6, 6))
plt.grid()
plt.title('v1 + v2\n')
plt.show()
```

### Vector * scalar

```python
v1 = np.array([3, -1])
lambda_ = -.3
v1m = v1 * lambda_

plt.plot((0, v1[0]), (0, v1[1]), 'b', label='v1')
plt.plot((0, v1m[0]), (0, v1m[1]), 'r:', label='lambda * v1')
plt.legend()
plt.axis('square')
axlim = max([abs(max(v1)), abs(max(v1m))]) * 1.2  #dynamic axis lim
plt.axis((-axlim, axlim, -axlim, axlim))
plt.grid()
plt.title('lambda * v1')
plt.show()
```

### Vector dot product

```python
# many ways to compute the dot product

v1 = np.array([1,  2, 3,  4, 5, 6])
v2 = np.array([0, -4, 1, -3, 6, 5])

dp1 = sum(np.multiply(v1, v2))

dp2 = np.dot(v1, v2)

dp3 = np.matmul(v1, v2)

dp4 = 0
for i in range(len(v1)):
    dp4 += v1[i] * v2[i]
    
print(dp1, dp2, dp3, dp4)
```

### Dot product properties

```python
# create 3 random vectors
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)
```

```python
# distributive
res1 = np.dot(a, (b+c))
res2 = np.dot(a, b) + np.dot(a, c)

print(res1, res2)
```

```python
# not associative
# one zeros vector
# a==b==c

# create 3 random vectors
n = 5
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)


bc = np.dot(b, c)
res1 = np.dot(a, bc)

ab = np.dot(a, b)
res2 = np.dot(ab, c)

print(res1, res2, sep='\n')
```

### Dot products with maxtrix columns

```python
# dot products with maxtrix columns
rows = 2

maxtrix_1 = []
maxtrix_2 = []
res = []

for i in range(3):
    maxtrix_1.append(np.random.randn(rows))
    maxtrix_2.append(np.random.randn(rows))
    col_wise_dot = np.dot(maxtrix_1[i], maxtrix_2[i])
    res.append(col_wise_dot)
    
print(res)
```

### Vector length

```python
# pythagorean theorem or dot product itself
from math import sqrt
v1 = np.array([2, 3, 7, 8])
v1_len = sqrt( np.dot(v1, v1) )
print(v1_len)
```

```python
v1 = np.array([2, 3, 7, 8])
v1_len = sqrt( sum(np.multiply(v1, v1)) )
print(v1_len)
```

```python
v1 = np.array([2, 3, 7, 8])
v1_len = np.linalg.norm(v1)
print(v1_len)
```
